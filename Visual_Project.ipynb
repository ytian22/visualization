# 1. Exploration

### 1. Data Type

```python
for i in data.columns:
    print i,": ",data.loc[:,i].dtype, len(data.loc[:,i].value_counts()),'\n'
```

### 2. Nulls

```python
check_nulls = data.isnull().sum()
check_nulls[check_nulls.values>0]
```

### 3. Categorize Columns

```python
for n,c in data.iteritems():
    if is_string_dtype(c):
        data[n] = c.astype('category').cat.as_ordered()
    elif len(data[n].value_counts())<100:
        data[n] = c.astype('category').cat.as_ordered()
```

```python
for i in data.columns:
    print i,": ",data.loc[:,i].dtype
```

### 4. Time

```python
data['issued_year'] = data.issue_d.str[-4:]
```

```python
data['issued_month']= data.issue_d.str[0:3]
```

### 5. Look at categorical columns

```python
for i in data.columns:
    if len(data.loc[:,i].value_counts()) < 10:
        print i, len(data.loc[:,i].value_counts())
```

```python
for i in data.columns:
    if (len(data.loc[:,i].value_counts()) > 10) & (len(data.loc[:,i].value_counts())<20):
        print i, len(data.loc[:,i].value_counts())
```
